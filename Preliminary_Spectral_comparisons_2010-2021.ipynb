```python
local_dataset_2010='moss_landing_2010.nc'
local_dataset_2011 = 'moss_landing_2011.nc'
local_dataset_2012='moss_landing_2012.nc'
local_dataset_2013='moss_landing_2013.nc'
local_dataset_2014='moss_landing_2014.nc'
local_dataset_2015='moss_landing_2015.nc'
local_dataset_2016='moss_landing_2016.nc'
local_dataset_2017 = 'moss_landing_2017.nc'
local_dataset_2018='moss_landing_2018.nc'
local_dataset_2019='moss_landing_2019.nc'
local_dataset_2021='moss_landing_2021.nc'
```

```python
ds_2010= xr.open_dataset(local_dataset_2010)
ds_2011=xr.open_dataset(local_dataset_2011)
ds_2012=xr.open_dataset(local_dataset_2012)
ds_2013=xr.open_dataset(local_dataset_2013)
ds_2014= xr.open_dataset(local_dataset_2014)
ds_2015=xr.open_dataset(local_dataset_2015)
ds_2016= xr.open_dataset(local_dataset_2016)
ds_2017=xr.open_dataset(local_dataset_2017)
ds_2018=xr.open_dataset(local_dataset_2018)
ds_2019=xr.open_dataset(local_dataset_2019)
ds_2021=xr.open_dataset(local_dataset_2021)
```

```python
df10=ds_2010.to_dataframe()
df11=ds_2011.to_dataframe()
df12=ds_2012.to_dataframe()
df13=ds_2013.to_dataframe()
df14=ds_2014.to_dataframe()
df15=ds_2015.to_dataframe()
df16=ds_2016.to_dataframe()
df17=ds_2017.to_dataframe()
df18=ds_2018.to_dataframe()
df19=ds_2019.to_dataframe()
df21=ds_2021.to_dataframe()
```

```python
ii_df10=np.array((df10['temp_flg'] == 1))
ii_df11=np.array((df11['temp_flg'] == 1)) 
ii_df12=np.array((df12['temp_flg'] == 1))
ii_df13=np.array((df13['temp_flg'] == 1))
ii_df14=np.array((df14['temp_flg'] == 1))
ii_df15=np.array((df15['temp_flg'] == 1))
ii_df16=np.array((df16['temp_flg'] == 1))
ii_df17=np.array((df17['temp_flg'] == 1)) 
ii_df18=np.array((df18['temp_flg'] == 1))
ii_df19=np.array((df19['temp_flg'] == 1))
ii_df21=np.array((df21['temp_flg'] == 1))
```

```python
temp_df10= np.array(df10['temp'][ii_df10])
temp_df11= np.array(df11['temp'][ii_df11])
temp_df12= np.array(df12['temp'][ii_df12])
temp_df13= np.array(df13['temp'][ii_df13])
temp_df14= np.array(df14['temp'][ii_df14])
temp_df15= np.array(df15['temp'][ii_df15])
temp_df16= np.array(df16['temp'][ii_df16])
temp_df17= np.array(df17['temp'][ii_df17])
temp_df18= np.array(df18['temp'][ii_df18])
temp_df19= np.array(df19['temp'][ii_df19])
temp_df21= np.array(df21['temp'][ii_df21])
```

```python
spectral_df10 = pd.DataFrame(temp_df10, columns=['temp']) 
spectral_df11 = pd.DataFrame(temp_df11, columns=['temp']) 
spectral_df12 = pd.DataFrame(temp_df12, columns=['temp']) 
spectral_df13 = pd.DataFrame(temp_df13, columns=['temp']) 
spectral_df14 = pd.DataFrame(temp_df14, columns=['temp']) 
spectral_df15 = pd.DataFrame(temp_df15, columns=['temp']) 
spectral_df16 = pd.DataFrame(temp_df16, columns=['temp']) 
spectral_df17 = pd.DataFrame(temp_df17, columns=['temp']) 
spectral_df18 = pd.DataFrame(temp_df18, columns=['temp']) 
spectral_df19 = pd.DataFrame(temp_df19, columns=['temp']) 
spectral_df21 = pd.DataFrame(temp_df21, columns=['temp']) 
```

```python
from scipy.signal import periodogram
from scipy.signal import welch

f,S = periodogram(spectral_df10['temp'],fs=288)
plt.figure()
plt.loglog(f,S,'k-')
plt.xlabel('[cpd]')
plt.ylabel('($^\circ$C)2')#variance per unit frequency 
plt.title('[periodogram, temp for 2010]')
N = len(spectral_df10['temp'])
fs = 288 
fN = 0.5* fs 
#####################################################################
f, S = welch(spectral_df10['temp'], fs=288, nperseg=N)

plt.figure()
plt.loglog(f,S)
plt.title('[Periodogram, temp for 2010]')
plt.xlabel('[cpd]')
plt.ylabel('($^\circ$C)2')#variance per unit frequency 
winlen = N/2
f2,S2 = welch (spectral_df10['temp'], fs=288., nperseg=winlen, window='hanning') 
plt.ylim(.0001) 
plt.xlim(.01)
```
